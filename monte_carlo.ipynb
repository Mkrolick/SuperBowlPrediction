```python
# afc.csv dataset

import pandas as pd
import numpy as np
import random

# Read the data
df = pd.read_csv('afc.csv')
df["OSRS"] = df["OSRS"].apply(lambda x: (x - np.mean(df[:]["OSRS"]))/np.std(df[:]["OSRS"]))
df["DSRS"] = df["DSRS"].apply(lambda x: (x - np.mean(df[:]["DSRS"]))/np.std(df[:]["DSRS"]))



teams = []
for i in range(len(df)):
    offensive = df.iloc[i]['OSRS']
    defensive = df.iloc[i]['DSRS']
    team = df.iloc[i]['Tm']
    
    teams += [[team, offensive, defensive]]



simulations = []
for i in range(10000):
    # create a list of 16 games
    games = []
    for j in range(16):
        # pick 2 teams randomly
        team1 = random.choice(teams)
        team2 = random.choice(teams)
        # calculate the score
        score1 = team1[1]*0.93 + team2[2]
        score2 = team2[1]*0.93 + team1[2]
        # add the game to the list
        games.append([team1[0], team2[0], score1, score2])
    # add the list of games to the list of simulations
    simulations.append(games)


# create a dictionary of teams
teams_dict = {}
for full_team in teams:
    teams_dict[full_team[0]] = 0

# loop through the simulations
for simulation in simulations:
    # loop through the games
    for game in simulation:
        # check if team1 won
        if game[2] > game[3]:
            teams_dict[game[0]] += 1
        # check if team2 won
        elif game[2] < game[3]:
            teams_dict[game[1]] += 1
        # check if it was a tie
        else:
            teams_dict[game[0]] += 0.5
            teams_dict[game[1]] += 0.5

    
# 4. print the results

print('Ranking of the teams:')
for team in sorted(teams_dict, key=teams_dict.get, reverse=True):
    print(team, teams_dict[team])
```
